```python
import random
def list_of_hexa_colors(num_colors):
    hexa_colors = []
    for _ in range(num_colors):
        hexa_color = '#{:06x}'.format(random.randint(0, 256**3-1))
        hexa_colors.append(hexa_color)
    return hexa_color
print(list_of_hexa_colors)
```

```python
import random

def generate_colors(num_colors, format):
    colors = []
    for _ in range(num_colors):
        if format == 'rgb':
            r = random.randint(0, 255)
            g = random.randint(0, 255)
            b = random.randint(0, 255)
            colors.append((r, g, b))
        elif format == 'hex': 
            color = '#{:06x}'.format(random.randint(0, 256**3-1))
            colors.append(color)
            return color
rgb_colors = generate_colors(5, 'rgb')
print(rgb_colors)

hex_colors = generate_colors(5, 'hex')
print(hex_colors)
```

```python
import random
 
def generate_unique_numbers():
    numbers = []
    while len(numbers) < 7:
       new_number = random.randint(0, 9)
       if new_number not in numbers:
          numbers.append(new_number)
          return numbers
```
